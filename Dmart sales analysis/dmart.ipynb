```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('DMart_sample_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.upper()  # removes spaces, lowercases everything
df.columns
```

```python
df = df.drop_duplicates()
df = df.fillna(0)
df['DATE'] = pd.to_datetime(df['DATE'], dayfirst=True)   # Fix date dtype
df['TIME'] = pd.to_datetime(df['TIME'], format='%H:%M:%S').dt.time
df.columns = df.columns.str.strip().str.upper()           # Already done ✅
df.head(20)
```

```python
df['MONTH'] = df['DATE'].dt.month_name()
df['YEAR'] = df['DATE'].dt.year
df['DAY_OF_WEEK'] = df['DATE'].dt.day_name()
df['HOUR'] = pd.to_datetime(df['TIME'], format='%H:%M:%S').dt.hour
df['REVENUE_PER_UNIT'] = df['TOTAL'] / df['QUANTITY']
df.head()
```

```python
# Same as above but cleaner syntax
df.query("CUSTOMERTYPE == 'Premium'")
df.query("TOTAL > 1000")
df.query("CUSTOMERTYPE == 'Premium' and TOTAL > 1000")
df.query("PRODUCTCATEGORY == 'Grocery' or PRODUCTCATEGORY == 'Electronics'")

# Using 'in' for multiple values
#df.query("PRODUCTCATEGORY in ['Grocery', 'Electronics']")
```

```python
df.query("YEAR =='2023'")
df.query("MONTH == 'August' ") 
df.query("CUSTOMERTYPE == 'Regular' ")
```
